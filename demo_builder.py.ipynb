```python
import ForceField as build
```

```python
# Non-Polarizable Force Field
ff = build.ForceField(['spc', 'Na', 'Cl'])
ff
```

```python
# Print To File
ff.print('ff.spce-Na-Cl.lmp')

# And to screen
ff.print()
```

```python
# Polarizable Drude Force Field
pol_ff = build.ForceField(['swm4-ndp', 'polNa', 'CO3'], polarizable=True)
pol_ff
```

```python
pol_ff.print('ff.swm4-ndp-Na-CO3.lmp')
pol_ff.print()
```

```python
pol_NaCl = build.ForceField(['swm4-ndp', 'polNa', 'polCl'], polarizable=True)
pol_NaCl.print('ff.swm4-ndp-Na-Cl.lmp')
pol_NaCl.print()
```

```python
pol_NaCl_asymmetric = build.ForceField(['swm4-ndp', 'polNa', 'polCl'], polarizable=True, symmetrized=False)
pol_NaCl_asymmetric.print('ff.asym.swm4-ndp-Na-Cl.lmp')
pol_NaCl_asymmetric.print()
```
